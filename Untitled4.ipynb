```python
def cross_val_smote(X, y, cv_n_splits = 5, smote_random_state=1, smote_kind = 'borderline-1', resample = 'true'):
    train_idx = []
    test_idx = []
    X_trains = []
    X_trains_smote = []
    X_tests = []
    y_trains = []
    y_trains_smote = []
    y_tests = []

    cv = StratifiedKFold(n_splits = cv_n_splits)
    smote = BorderlineSMOTE(random_state = smote_random_state, kind = smote_kind)
  
    for train, test in cv.split(X,y.ravel()):
        train_idx.append(train)
        test_idx.append(test)
    
    for item in train_idx:
        X_tmp = []
        y_tmp = []
        for index in item:
            X_tmp.append(X[index])
            y_tmp.append(y[index])
        X_trains.append(X_tmp)
        y_trains.append(y_tmp)
    
    for item in test_idx:
        X_tmp = []
        y_tmp = []
        for index in item:
            X_tmp.append(X[index])
            y_tmp.append(y[index])
        X_tests.append(X_tmp)
        y_tests.append(y_tmp)
        
    if resample == 'false':
        return X_trains, X_tests, y_trains, y_tests
    if resample == 'true':
        for i in range(0, len(X_trains)):
            X_smt, y_smt = smote.fit_resample(X_trains[i], y_trains[i])
            X_trains_smote.append(X_smt)
            y_trains_smote.append(y_smt)
        return X_trains_smote, X_tests, y_trains_smote, y_tests
```

```python
X_trains, X_tests, y_trains, y_tests = cross_val_smote(X, y, resample = 'false')
```

```python
X_trains_smote, X_tests, y_trains_smote, y_tests = cross_val_smote(X, y, resample = 'true')
```

### Select K for KNN

```python
Ks = np.arange(11,21,1)
precisions = []
auprcs = []
mean_recall = np.linspace(0,1,100)
plt.figure(figsize=(20,20))
colors = cm.rainbow(np.linspace(0, 1, 10))
for K, c in zip(Ks, colors):
    KNN = KNeighborsClassifier(n_neighbors=K)
    for X_train_smote, X_test, y_train_smote, y_test in zip(X_trains_smote, X_tests, y_trains_smote, y_tests):
        X_train_smote = np.array(X_train_smote)
        X_test = np.array(X_test)
        y_train_smote = np.array(y_train_smote)
        y_test = np.array(y_test)
        prediction = KNN.fit(X_train_smote, y_train_smote.ravel()).predict_proba(X_test)
        precision, recall, threshhold = precision_recall_curve(y_test.ravel(), prediction[:, 1])
        precisions.append(np.interp(mean_recall, precision, recall))
        auprc = auc(recall, precision)
        auprcs.append(auprc) 
        plt.plot(recall, precision, lw=2, alpha=0.2)
    mean_precision = np.mean(precisions, axis=0)
    mean_auprc = auc(mean_recall, mean_precision)
    plt.plot(mean_precision, mean_recall, color = c, 
             label=r'K={} (AUPRC = {:.2f})'.format(K, mean_auprc), lw=2, alpha = 1)
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Area Under Precision-Recall Curve')
plt.legend(loc="lower left")
plt.show()
    
```

```python
def plot_cross_val_PR_curve(models, colors, names, X, y, baseline_model=ComplementNB(), smote='false'):
    X_trains, X_tests, y_trains, y_tests = cross_val_smote(X, y, resample = 'false')
    precisions = []
    auprcs = []
    mean_recall = np.linspace(0,1,100)
    precisions_baseline = []
    auprcs_baseline = []
    mean_recall_baseline = np.linspace(0,1,100)
    plt.figure(figsize=(20,20))
    if smote == 'false':
        for model, color, name in zip(models, colors, names):
            for X_train, X_test, y_train, y_test in zip(X_trains, X_tests, y_trains, y_tests):
                X_train = np.array(X_train)
                X_test = np.array(X_test)
                y_train = np.array(y_train)
                y_test = np.array(y_test)
                prediction = model.fit(X_train, y_train.ravel()).predict_proba(X_test)
                precision, recall, threshhold = precision_recall_curve(y_test.ravel(), prediction[:, 1])
                precisions.append(np.interp(mean_recall, precision, recall))
                auprc = auc(recall, precision)
                auprcs.append(auprc) 
                plt.plot(recall, precision, lw=2, alpha=0.2)
            mean_precision = np.mean(precisions, axis=0)
            mean_auprc = auc(mean_recall, mean_precision)
            plt.plot(mean_precision, mean_recall, color = color,
                     label=r'{} Classifier (AUPRC = {:.2f})'.format(name, mean_auprc), lw=2, alpha = 1)
            
        for X_train, X_test, y_train, y_test in zip(X_trains, X_tests, y_trains, y_tests):
            X_train = np.array(X_train)
            X_test = np.array(X_test)
            y_train = np.array(y_train)
            y_test = np.array(y_test)
            prediction = baseline_model.fit(X_train, y_train.ravel()).predict_proba(X_test)
            precision, recall, threshhold = precision_recall_curve(y_test.ravel(), prediction[:, 1])
            precisions_baseline.append(np.interp(mean_recall_baseline, precision, recall))
            auprc = auc(recall, precision)
            auprcs_baseline.append(auprc) 
        mean_precision_baseline = np.mean(precisions_baseline, axis=0)
        mean_auprc = auc(mean_recall_baseline, mean_precision_baseline)
        plt.plot(mean_precision_baseline, mean_recall_baseline, color = 'black',
                 label=r'Baseline Model: ComplementNB (AUPRC = {:0.2f})'.format(mean_auprc), lw=2, alpha = 1)
        plt.xlabel('Recall')
        plt.ylabel('Precision')
        plt.title('Area Under Precision-Recall Curve with original data')
        plt.legend(loc="lower left")
        plt.show()
    if smote == 'true':
        X_trains_smote, X_tests, y_trains_smote, y_tests = cross_val_smote(X, y, resample = 'true')
        for model, color, name in zip(models, colors, names):
            for X_train_smote, X_test, y_train_smote, y_test in zip(X_trains_smote, X_tests, y_trains_smote, y_tests):
                X_train_smote = np.array(X_train_smote)
                X_test = np.array(X_test)
                y_train_smote = np.array(y_train_smote)
                y_test = np.array(y_test)
                prediction = model.fit(X_train_smote, y_train_smote.ravel()).predict_proba(X_test)
                precision, recall, threshhold = precision_recall_curve(y_test.ravel(), prediction[:, 1])
                precisions.append(np.interp(mean_recall, precision, recall))
                auprc = auc(recall, precision)
                auprcs.append(auprc) 
                plt.plot(recall, precision, lw=2, alpha=0.2)
            mean_precision = np.mean(precisions, axis=0)
            mean_auprc = auc(mean_recall, mean_precision)
            plt.plot(mean_precision, mean_recall, color = color,
                     label=r'{} Classifier (AUPRC = {:.2f})'.format(name, mean_auprc), lw=2, alpha = 1)
            
        for X_train, X_test, y_train, y_test in zip(X_trains, X_tests, y_trains, y_tests):
            X_train = np.array(X_train)
            X_test = np.array(X_test)
            y_train = np.array(y_train)
            y_test = np.array(y_test)
            prediction = baseline_model.fit(X_train, y_train.ravel()).predict_proba(X_test)
            precision, recall, threshhold = precision_recall_curve(y_test.ravel(), prediction[:, 1])
            precisions_baseline.append(np.interp(mean_recall_baseline, precision, recall))
            auprc = auc(recall, precision)
            auprcs_baseline.append(auprc) 
        mean_precision_baseline = np.mean(precisions_baseline, axis=0)
        mean_auprc = auc(mean_recall_baseline, mean_precision_baseline)
        plt.plot(mean_precision_baseline, mean_recall_baseline, color = 'black',
                 label=r'Baseline Model: ComplementNB (AUPRC = {:.2f})'.format(mean_auprc), lw=2, alpha = 1)
        plt.xlabel('Recall')
        plt.ylabel('Precision')
        plt.title('Area Under Precision-Recall Curve with SMOTE-Resampling')
        plt.legend(loc="lower left")
        plt.show()
```
